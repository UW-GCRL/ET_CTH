```python
import netCDF4 as nc
import numpy as np

# 设置数据的路径和文件名模板
base_path = 'G:\\Hangkai\\CTH_ET project\\simulated data\\Default_0.8_case\\archive\\Default_0.8_case\\lnd\\hist\\'
file_template = 'Default_0.8_case.clm2.h0.{}-02-01-00000.nc'

# 用于存储平均数据的字典
avg_data = {}

# 循环读取2010年到2014年的数据
for year in range(2010, 2015):
    file_path = base_path + file_template.format(year)
    with nc.Dataset(file_path, 'r') as ds:
        # 只在第一次迭代时获取变量列表
        if year == 2010:
            variables = list(ds.variables.keys())

        # 对每个变量进行处理
        for var_name in variables:
            if var_name not in avg_data:
                avg_data[var_name] = ds.variables[var_name][:]
            else:
                avg_data[var_name] += ds.variables[var_name][:]

# 计算平均值
for var_name in avg_data:
    avg_data[var_name] /= 5

# 创建新的NetCDF文件来存储平均数据
with nc.Dataset('G:\\Hangkai\\CTH_ET project\\simulated data\\Default_0.8_average_data_2001_2014.nc', 'w') as new_ds:
    # 使用最后一次读取的原始数据文件作为模板复制维度和属性
    with nc.Dataset(file_path, 'r') as ds:
        # 复制维度
        for name, dim in ds.dimensions.items():
            new_ds.createDimension(name, len(dim) if not dim.isunlimited() else None)

        # 复制并存储变量
        for var_name, data in avg_data.items():
            # 根据原文件创建变量
            original_var = ds.variables[var_name]
            new_var = new_ds.createVariable(var_name, original_var.datatype, original_var.dimensions)
            new_var.setncatts({attr: original_var.getncattr(attr) for attr in original_var.ncattrs()})
            new_var[:] = data

print("Average data saved!")
```

```python
import netCDF4 as nc
import numpy as np

# 从2010年开始到2014年 四年的平均！
# 2001年到2009年的九年都是初始化

# 设置数据的路径和文件名模板
base_path = 'G:\\Hangkai\\CTH_ET project\\simulated data\\Default_0.8_case\\archive\\Default_0.8_case\\lnd\\hist\\'
file_template = 'Default_0.8_case.clm2.h0.{}-02-01-00000.nc'

# 初始化存储所有变量数据的字典
data_dict = {}
data_types = {}

# 读取第一个年份的文件来初始化data_dict
first_ds = nc.Dataset(base_path + file_template.format(2010), 'r')

with nc.Dataset(base_path + file_template.format(2010), 'r') as ds:
    for var_name in ds.variables:
        if var_name not in ['time', 'time_bounds', 'date_written', 'time_written']:  # 排除某些不需要计算平均值的变量
            data_dict[var_name] = ds.variables[var_name][:]
            data_types[var_name] = ds.variables[var_name].dtype

# 循环读取2001年到2014年的数据
for year in range(2011, 2015):  # 从2011开始，因为2001年的数据已经在上面读取了
    file_path = base_path + file_template.format(year)
    with nc.Dataset(file_path, 'r') as ds:
        for var_name in data_dict:
            data_dict[var_name] += ds.variables[var_name][:]

# 计算每个变量的平均值
for var_name in data_dict:
    # 先将数组转换为浮点类型
    data_dict[var_name] = data_dict[var_name].astype(np.float64)
    data_dict[var_name] /= 5
    data_dict[var_name] = data_dict[var_name].astype(data_types[var_name])

# 创建新的NetCDF文件来存储平均数据
with nc.Dataset('G:\\Hangkai\\CTH_ET project\\simulated data\\Default_0.8_average_data_2001_2014.nc', 'w') as new_ds:
    # 使用first_ds文件的维度来创建新文件的维度
    for name, dim in first_ds.dimensions.items():
        new_ds.createDimension(name, len(dim) if not dim.isunlimited() else None)
    
    # 写入每个变量的平均数据
    for var_name, avg_value in data_dict.items():
        # 提取_FillValue属性（如果存在的话）
        fill_value = None
        if '_FillValue' in first_ds.variables[var_name].ncattrs():
            fill_value = getattr(first_ds.variables[var_name], '_FillValue')
        
        new_var = new_ds.createVariable(var_name, data_types[var_name], first_ds.variables[var_name].dimensions, fill_value=fill_value)
        new_var[:] = avg_value
        
        # 复制原变量的属性到新变量
        for attr_name in first_ds.variables[var_name].ncattrs():
            if attr_name != '_FillValue':  # 已经在变量创建时处理了
                setattr(new_var, attr_name, getattr(first_ds.variables[var_name], attr_name))

first_ds.close()
print("Average data saved!")
```
