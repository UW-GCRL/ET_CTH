```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import rasterio

%matplotlib inline
```

```python
def plot_geotiff(file_path, title, ax, cmap='YlGn', add_coastlines=True, central_longitude=180):
    with rasterio.open(file_path) as src:
        data = src.read(1)
        transform = src.transform
        extent = [transform[2], transform[2] + src.width * transform[0], transform[5] + src.height * transform[4], transform[5]]
        
        cmap = plt.get_cmap(cmap)
        cmaplist = [cmap(i) for i in range(cmap.N)]
        cmaplist[0] = (1, 1, 1, 1.0)  # Set the color for value 0 to white
        custom_cmap = mcolors.LinearSegmentedColormap.from_list('Custom cmap', cmaplist, cmap.N)
        
        img = ax.imshow(data, cmap=custom_cmap, extent=extent, transform=ccrs.PlateCarree(), origin='upper')
        
        if add_coastlines:
            ax.coastlines()
        
        ax.set_title(title,fontsize=9)
        ax.set_global()
        ax.set_extent([-180, 180, -90, 90], crs=ccrs.PlateCarree(central_longitude=central_longitude))
```

```python
import io
import rasterio
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import cartopy.crs as ccrs
from PIL import Image

fig, axes = plt.subplots(dpi=400, nrows=2, ncols=2, figsize=(20, 12), subplot_kw={'projection': ccrs.PlateCarree()})

plot_geotiff(r'H:\CTH_final_map\final_map\GLAS_GLOBAL.tif', 'CLM default', axes[0, 0])
plot_geotiff(r'H:\CTH_final_map\final_map\FinalMax2020.tif', 'GEDI max', axes[0, 1])
plot_geotiff(r'H:\CTH_final_map\final_map\FinalMean2020.tif', 'GEDI mean', axes[1, 0])
plot_geotiff(r'H:\CTH_final_map\final_map\FinalMedian2020.tif', 'GEDI median', axes[1, 1])

plt.tight_layout()

# Add a single colorbar for all subplots
cax = fig.add_axes([0.93, 0.125, 0.03, 0.755])
sm = plt.cm.ScalarMappable(cmap='YlGn', norm=plt.Normalize(vmin=0, vmax=40))
sm.set_array([])
fig.colorbar(sm, cax=cax)

# Save the figure to a buffer in memory
fig.tight_layout()
buf = io.BytesIO()
plt.savefig(buf, format='tiff', dpi=200)
buf.seek(0)

# Read the buffer as an image and save it as a TIFF file
im = Image.open(buf)
im.save('canopy_top_height.tiff', 'TIFF')

# Close the buffer
buf.close()

plt.show()
```

```python
import io
import rasterio
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import cartopy.crs as ccrs
from PIL import Image

def plot_geotiff(file_path, title, ax, cmap='YlGn', add_coastlines=True, central_longitude=180, norm=plt.Normalize(vmin=0, vmax=40)):
    with rasterio.open(file_path) as src:
        data = src.read(1)
        transform = src.transform
        extent = [transform[2], transform[2] + src.width * transform[0], transform[5] + src.height * transform[4], transform[5]]
        
        cmap = plt.get_cmap(cmap)
        cmaplist = [cmap(i) for i in range(cmap.N)]
        cmaplist[0] = (1, 1, 1, 1.0)  # Set the color for value 0 to white
        custom_cmap = mcolors.LinearSegmentedColormap.from_list('Custom cmap', cmaplist, cmap.N)
        
        img = ax.imshow(data, cmap=custom_cmap, extent=extent, transform=ccrs.PlateCarree(), origin='upper', norm=norm)
        
        if add_coastlines:
            ax.coastlines()
        
        ax.set_title(title)
        ax.set_global()
        ax.set_extent([-180, 180, -90, 90], crs=ccrs.PlateCarree(central_longitude=central_longitude))

fig, axes = plt.subplots(dpi=400, nrows=2, ncols=2, figsize=(20, 12), subplot_kw={'projection': ccrs.PlateCarree()})

plot_geotiff(r'H:\CTH_final_map\final_map\GLAS_GLOBAL.tif', 'CLM default', axes[0, 0])
plot_geotiff(r'H:\CTH_final_map\final_map\FinalMax2020.tif', 'GEDI max', axes[0, 1])
plot_geotiff(r'H:\CTH_final_map\final_map\FinalMean2020.tif', 'GEDI mean', axes[1, 0])
plot_geotiff(r'H:\CTH_final_map\final_map\FinalMedian2020.tif', 'GEDI median', axes[1, 1])

plt.tight_layout()

# Add a single colorbar at the bottom of the figure
cax = fig.add_axes([0.125, 0.1, 0.755, 0.03])
sm = plt.cm.ScalarMappable(cmap='YlGn', norm=plt.Normalize(vmin=0, vmax=40))
sm.set_array([])
cbar = fig.colorbar(sm, cax=cax, orientation='horizontal')
cbar.set_label('Canopy height (m)')

# Save the figure to a buffer in memory
fig.tight_layout()
buf = io.BytesIO()
plt.savefig(buf, format='tiff', dpi=200)
buf.seek(0)

# Read the buffer as an image and save it as a TIFF file
im = Image.open(buf)
im.save('canopy_top_height_4_scenarios.tiff', 'TIFF')

# Close the buffer
buf.close()

plt.show()
```
